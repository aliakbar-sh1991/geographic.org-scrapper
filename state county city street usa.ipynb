```python
import requests
from bs4 import BeautifulSoup
import json
from google.colab import drive
drive.mount('/content/drive')
```

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
```

```python
main_Dict={}
request = requests.get('https://geographic.org/streetview/usa/',headers=headers)
soup= BeautifulSoup(request.text, 'html.parser')
for i in soup.find_all('span' , {'class':'listspan'})[0].find_all('a'):
  x=f'{i.get_text().strip()}'
  y=i.attrs['href'].split('/')[0]
  main_Dict[x]={}
  main_Dict[x]['st_code']=y
  main_Dict[x]['link']=f"https://geographic.org/streetview/usa/{i.attrs['href']}"
```

```python
for key1 in main_Dict.keys():
  lnk = f"https://geographic.org/streetview/usa/{main_Dict[key1]['st_code']}"
  request = requests.get(lnk,headers=headers)
  soup= BeautifulSoup(request.text, 'html.parser')
  try:
    for i in soup.find_all('div' , {'class':'listmain'})[0].find_all('a'):
      main_Dict[key1][i.get_text().strip()]={}
  except:
    pass
```

```python
for key1 in main_Dict.keys():
  for key2 in main_Dict[key1].keys():
    lnk = f"https://geographic.org/streetview/usa/{main_Dict[key1]['st_code']}/{key2.lower().replace(' ','')}/index.html"
    request = requests.get(lnk,headers=headers)
    soup= BeautifulSoup(request.text, 'html.parser')   
    if key2!='link' and key2!='st_code':
      try:
        for i in soup.find_all('div' , {'class':'listmain'})[0].find_all('a'):
          main_Dict[key1][key2][i.get_text().strip()]={}
      except:
        pass
```

```python
for key1 in main_Dict.keys():
  for key2 in main_Dict[key1].keys():   
    if key2!='link' and key2!='st_code':
      for key3 in main_Dict[key1][key2]:
        lnk = f"https://geographic.org/streetview/usa/{main_Dict[key1]['st_code']}/{key2.lower().replace(' ','')}/{key3.lower().replace(' ','')}.html"
        request = requests.get(lnk,headers=headers)
        soup= BeautifulSoup(request.text, 'html.parser')
        main_Dict[key1][key2][key3]['Streets']=[]
        try:
          for i in soup.find_all('div' , {'class':'listmain'})[0].find_all('li'):
            q={}
            q['name']=' '.join(i.get_text().split(' ')[0:-1]).strip()
            q['postal code']=i.get_text().split(' ')[-1]
            main_Dict[key1][key2][key3]['Streets'].append(q)
        except:
          pass
```

```python
x = json.dumps(main_Dict)
```
